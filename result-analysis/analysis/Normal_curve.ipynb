```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
lim=30
x = np.arange(-lim, lim, 0.001)
ax.plot(x, norm.pdf(x, -10, 8), color='blue', linewidth=2, linestyle='--')
ax.plot(x, norm.pdf(x, 10, 7), color='red', linewidth=2)
ax.plot(x, norm.pdf(x, -10, 3), color='red', linewidth=2)
ax.plot(x, norm.pdf(x, 10, 2), color='blue', linewidth=2, linestyle='--')
ax.set_xlim(-lim, lim)
ax.axis('off')
# ax.get_xaxis().set_visible(False)
# ax.get_yaxis().set_visible(False)
# ax.spines['top'].set_visible(False)
# ax.spines['right'].set_visible(False)
# ax.spines['left'].set_visible(False)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
lim=30
x = np.arange(-lim, lim, 0.001)
ax.plot(x, norm.pdf(x, -8, 8), color='red', linewidth=2)
ax.plot(x, norm.pdf(x, 8, 8), color='red', linewidth=2)
ax.plot(x, norm.pdf(x, -5, 5), color='blue', linewidth=2, linestyle='--')
ax.plot(x, norm.pdf(x, 5, 5), color='blue', linewidth=2, linestyle='--')
ax.set_xlim(-lim, lim)
ax.axis('off')
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
x0 = np.arange(0, 1/3, 0.001)
ax.plot(x0, x0, color='k', linewidth=2, linestyle='--')
x = np.arange(1/3, 1, 0.001)
ax.plot(x, x, color='k', linewidth=2, label='max P(predict correctly)')
ax.plot(x, x**2+((1-x)**2)/2, color='red', linewidth=2, label='min P(self-coherence)')
x1 = np.arange(1/3, 1/2, 0.001)
ax.plot(x1, 6*x1**2-4*x1+1, color='blue', linewidth=2, linestyle='--', label='max P(self-coherence)')
x2 = np.arange(1/2, 1, 0.001)
ax.plot(x2, x2**2+(1-x2)**2, color='blue', linewidth=2, linestyle='--')
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
# ax.axis('on')
# setting ticks for x-axis 
ax.set_xticks([0.333, 0.5, 1])
# setting label for x tick
ax.set_xticklabels(['1/3', '1/2', '1'], fontdict={'fontsize': 18, 'family':'Arial'})
# setting ticks for y-axis 
ax.set_yticks([0.333, 0.5, 1])
ax.set_yticklabels(['1/3', '1/2', '1'], fontdict={'fontsize': 18, 'family':'Arial'})
ax.set_xlabel('P(A)', fontdict={'fontsize': 18, 'family':'Arial', 'weight': 'bold'})
ax.legend(prop={'family':'Arial', 'size':17}, fancybox=True, loc='upper left')
```
